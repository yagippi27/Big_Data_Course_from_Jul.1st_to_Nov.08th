연습 문제 4.2.10¶
두 개의 데이터프레임을 만들고 merge 명령으로 합친다. 단 데이터프레임은 다음 조건을 만족해야 한다.

1. 각각 5 x 5 이상의 크기를 가진다.
2. 공통 열을 하나 이상 가진다. 다만 공통 열의 이름은 서로 다르다.

```python
df1 = pd.DataFrame({
    '학번': [201202857, 201503578, 201201285, 201300367, 201607861],
    '학점': [4.3, 4.0, 3.72, 3.45, 3.9],
    '학과': ['사회학과', '물리학과', '경영학과', '컴퓨터공학과', '영어영문학과'],
    '고향': ['서울', '대구', '대전', '서울', '경기도'],
    '이름': ['안뫄뫄', '박놔놔', '김화화', '최솨솨', '이과과']
    
}, columns = ['학번', '이름', '학과', '학점', '고향'])
df1
```

```python
df2 = pd.DataFrame({
    '학생번호': [201202857, 201503578, 201201285, 201300367, 201607861],
    '용돈금액' : [300000, 250000, 500000, 450000, 200000],
    '통학방법' : ["버스", "자전거", "자동차", "도보", "도보"],
    '거주지' : ["집", "집", "자취", "기숙사", "자취"],
    '코딩가능여부' : ["yes", "yes", "yes", "yes", "no"]
}, columns = ['학생번호', '용돈금액', '통학방법', '거주지', '코딩가능여부'])
df2
```

```python
pd.merge(df1, df2, left_on = "학번", right_on="학생번호")
```

연습 문제 4.2.11
어느 회사의 전반기(1월 ~ 6월) 실적을 나타내는 데이터프레임과 후반기(7월 ~ 12월) 실적을 나타내는 데이터프레임을 만든 뒤 합친다. 실적 정보는 "매출", "비용", "이익" 으로 이루어진다. (이익 = 매출 - 비용).

또한 1년간의 총 실적을 마지막 행으로 덧붙인다.

```python
df_전 = pd.DataFrame({
    "매출" : [3000, 4000, 4500, 5000, 7000, 6500],
    "비용" : [1500, 1700, 1800, 3000, 3000, 2500],
    "이익" : [1500, 2300, 2700, 2000, 4000, 4000]
}, index = ["1월", "2월", "3월", "4월", "5월", "6월"], columns = ["매출", "비용", "이익"])
df_전
```

```python
df_후 = pd.DataFrame({
    "매출" : [6500, 5500, 6500, 5000, 4000, 2500],
    "비용" : [3000, 2000, 2500, 2000, 1500, 1500]
}, index = ["7월", "8월", "9월", "10월", "11월", "12월"], columns = ["매출", "비용"])
df_후
```

```python
df_후["이익"] = df_후["매출"] - df_후["비용"]
df_후
```

```python
df_전후 = pd.concat([df_전, df_후])
df_rt= df_전후.T
df_rt
```

```python
df_rt["총실적"] = df_rt["1월"]+df_rt["2월"]+df_rt["3월"]+df_rt["4월"]+df_rt["5월"]+ \
df_rt["6월"]+df_rt["7월"]+df_rt["8월"]+df_rt["9월"]+df_rt["10월"]+df_rt["11월"]+df_rt["12월"]
```

```python
df_rt
```

```python
df_총실적 = df_rt.T
df_총실적
```

연습 문제 1¶
key1의 값을 기준으로 data1의 값을 분류하여 합계를 구한 결과를 시리즈가 아닌 데이터프레임으로 구한다.

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
groups = df2.groupby(df2.key1)
groups
```

```python
groups.groups
```

```python
pd.DataFrame(df2.data1.groupby(df2.key1).sum())
```

```python
iris.head(10)
```

```python
groups2 =iris.groupby(iris.species)
groups2.groups
```

```python
pd.DataFrame(iris.sepal_length.groupby(iris.species).mean())
```
