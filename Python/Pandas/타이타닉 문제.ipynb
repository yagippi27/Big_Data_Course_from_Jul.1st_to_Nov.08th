타이타닉호 승객을 사망자와 생존자 그룹으로 나누고 각 그룹에 대해 

'미성년자', '청년', '중년', '장년', '노년' 승객의 비율을 구한다. 

각 그룹 별로 비율의 전체 합은 1이 되어야 한다.

뒤의 노트에 작성~

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
titanic
```

```python
def select_alive (df):
    return df.alive == 'yes'
alive = titanic.loc[select_alive(titanic)]
```

```python
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
alive_ages = pd.cut(alive['age'], bins, labels=labels)

alive_ages = pd.DataFrame(alive_ages.value_counts(), index = labels)
alive_ages['비율'] =  alive_ages["age"]/(alive_ages['age'].sum())
alive_ages.loc['합계', :] = alive_ages.sum()
alive_ages
```

```python
notalive = titanic.loc[select_alive(titanic) == False]
notalive.head()
```

```python
notalive_ages = pd.cut(notalive['age'], bins, labels=labels)

notalive_ages = pd.DataFrame(notalive_ages.value_counts(), index = labels)
notalive_ages['비율'] =  notalive_ages["age"]/(notalive_ages['age'].sum())
notalive_ages.loc['합계', :] = notalive_ages.sum()
notalive_ages
```

```python
bins = [1, 15, 25, 35, 60, 99]
labels = ["미성년자", "청년", "중년", "장년", "노년"]
age = pd.cut(titanic["age"], bins, labels = labels)
age
```
