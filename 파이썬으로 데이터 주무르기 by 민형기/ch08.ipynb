## 워드 클라우드

```python
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open("D:/workspace/Python/Book/DataScience-master/data/09. alice.txt").read()
alice_mask = np.array(Image.open('D:/workspace/Python/Book/DataScience-master/data/09. alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['font.size'] = 10
%matplotlib inline
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
# cmap = color map -> 흑백 사진을 원할 때 gray라고 씀
plt.axis('off') # axis('on')을 하면 테두리가 생김
plt.show()
```

```python
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords = stopwords)
wc= wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## 육아휴직관련 법안에 대한 분석

```python
import nltk
```

```python
from konlpy.corpus import kobill
files_ko = kobill.fileids()
doc_ko = kobill.open('1809890.txt').read()
```

```python
doc_ko
```

```python
from konlpy.tag import Twitter; t = Twitter()
tokens_ko = t.nouns(doc_ko)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name = '대한민국 국회 의안 제 1809890호')
```

```python
print(len(ko.tokens))
print(len(set(ko.tokens)))
ko.vocab()
```

```python
plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
stop_words = ['.', '()', ')', ',', "'", '%', '-', 'X', ').', 'x', '의', '자', '에', '안', '번', '호', '을', '이', '다', '만', '로', '가', '를']
ko = [each_word for each_word in ko if each_word not in stop_words]
ko
```

```python
ko = nltk.Text(ko, name = '대한민국 국회 의안 제 1809890호')

plt.figure(figsize=(12,6))
ko.plot(50)
plt.show()
```

```python
ko.count('초등학교')
```

```python
plt.figure(figsize=(12,6))
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

```python
ko.concordance('초등학교')
```

```python
ko.collocations()
```

```python
data = ko.vocab().most_common(150)

wordcloud = WordCloud(font_path='C:\Windows/Fonts/HMKLP.ttf', relative_scaling = 0.2, background_color = 'white').generate_from_frequencies(dict(data))
plt.figure(figsize=(12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
